## Sobre el dataset

En el dataset provisto por la cátedra se encuentran 53127 grabaciones de latidos. Cada látido tiene su anotación especificando que tipo de latido es. Los tipos de latidos estan identificados de la siguiente forma:

- Normales, identificados con el número 0.

- Ventriculares prematuros, identificados con el número 1.

- Nodales prematuros, identificados con el número 2.

- Supraventriculares prematuros, identificados con el número 3.

El látido anotado es el que tiene el pico en el medio de la serie temporal, la misma tiene parte del latido anterior y el latidos posterior al anotado. La serie puede ser recortada de la forma que resulte conveniente para el procesamiento de los datos, es decir, si se observa que tomando n puntos (siendo $ n < tamaño /2$) a la izquierda y derecha del medio de la medición se obtiene un mejor rendimiento 

El dataset contiene 47000 latidos normales, 5206 latidos ventriculares prematuros, 119 latidos nodales prematuros y 802 latidos supraventriculares prematuros.

Se puede dividir el dataset en el set de datos de entrenamiento y set de datos de testeo de la manera que se considere mejor.

A continuación se mostrará como acceder al dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

lats = np.load('train_lats_part1.npy')
anotaciones = np.load('train_anot.npy')

nr_samples = lats.shape[0]
nr_classes = 4

#se tienen dos señales por cada látido tomadas de diferentes lugares en el cuerpo
plt.figure()
plt.subplot(211)
plt.title('Latido Normal')
plt.plot(lats[1,0])
plt.subplot(212)
plt.plot(lats[1,1])

        
plt.figure()
plt.subplot(211)
plt.title('Latido Ventricular Prematuro')
plt.plot(lats[17,0])
plt.subplot(212)
plt.plot(lats[17,1])


plt.figure()
plt.subplot(211)
plt.title('Latido Nodales Prematuro')
plt.plot(lats[122,0])
plt.subplot(212)
plt.plot(lats[122,1])

plt.figure()
plt.subplot(211)
plt.title('Latido Supraventricular Prematuro')
plt.plot(lats[98,0])
plt.subplot(212)
plt.plot(lats[98,1])

plt.show()

```

To classify ECG signals into different classes of cardiac functioning, one should think about the features that might distinguish each class. From signal theory it is known that the Fourier Transform shows the spectral information of a signal. This spectral information can be really different for signals that might look similar in time-domain. Also the signal's mean and variance are properties that can be used to describe a signal. For these reasons, I chose to create an input vector for my network that consists of the original signal (the two leads), the positive part of the FFT, the mean and the variance of the signal. Initially I tested without adding the mean and variance, but adding these two features to the input vector mainly improved the classification of the fourth class (Latido Supraventricular prematuro). 


```python
def fft_one_sided(S):
    fft = np.fft.fft(S)
    fft_pos = fft[200:]
    return abs(fft_pos)

#Initialize the training data
data = np.zeros((nr_samples,1202))
new_fft = np.zeros(400)

# The input vector is saved in a numpy array of size nr_samples x 1202.
# The first 400 scalars of the vector are the FFTs of the two leads. Since only the positive parts of the FFTs are used, each FFT is 200 samples.
# Scalars 400-800 is the original signal from lead 1.
# Scalars 800-1200 is the original signal from lead 2
# Inputs 1201 and 1202 are respectively the mean and variance of the signal. 
for i in range(0,nr_samples):
    #Process both leads
    for j in range(0,2):
        signal = lats[i,j]
        
        #Fill the first part of the input vector with the ffts
        start_point = 200*j  #either fill the array from 0 to 200 or 200 to 400 (dependent on the lead)
        end_point = start_point+200
        new_fft[start_point:end_point] = fft_one_sided(signal)
        
        #Add the original data signal leads also to the input
        data[i,start_point+400+(j*200):end_point+400+(j+1)*200] = signal

    data[i,0:400] = new_fft
                      
    #Compute signal statistics to add to the input vector as well
    data[i,1200] = np.mean(signal)
    data[i,1201] = np.var(signal)
                      
    
#Target data has to be set as one-hot encoding
targets = np.zeros((nr_samples,nr_classes))

for i in range(0,nr_samples):
    targets[i,int(anotaciones[i])] = 1
```

The dataset has to be split into training, validation and test data. I chose to reserve 20 percent for validation, use 10 percent for testing, and the remaining 70 percent for training.

```python
#Split the data intro train, validation and test set
val_perc = 0.2
test_perc = 0.1

nr_val_data = int(np.floor(val_perc * nr_samples))
nr_test_data = int(np.floor(test_perc * nr_samples))
nr_train_data = int(nr_samples - nr_val_data - nr_test_data)

print('Number of samples in the training, validation and testing set are respectively :', nr_train_data, ', ', nr_val_data, ' and ', nr_test_data)
 
#Initialize numpy arrays for holding the train,validation and test data and targets
train_data = np.zeros((nr_train_data,400))
train_target = np.zeros(nr_train_data)
val_data = np.zeros((nr_val_data,40))
val_target = np.zeros(nr_val_data)
test_data = np.zeros((nr_test_data,400))
test_target = np.zeros(nr_test_data)

#Divide the data (and the corresponding targets) in training, validation and testing set
train_data = data[0:nr_train_data,:]
train_target = targets[0:nr_train_data,:]

val_data = data[nr_train_data:(nr_train_data + nr_val_data),:]
val_target = targets[nr_train_data:(nr_train_data + nr_val_data),:]

test_data = data[(nr_train_data + nr_val_data):nr_samples,:]
test_target = targets[(nr_train_data + nr_val_data):nr_samples,:]
```

Signals of classes 0 and 1 are represented much more in the database, than the signals from classes 2 and 3. When running networks without changing anything about this, it resulted in the fact that the network had a bias towards classifying signals as type 0 (normal). 
Since the penalty for classifying an arythmetric as a normal signal is really high, I wanted to improve this bias. For this reason, I up-sampled class 2 and 3 in the training set. Up-sampling was done by just copying signals and it's targets.
The amount of up-sampling that could be used for class 2 and 3 was free to choose. I empirically determined to use multipliers of respectively 20 and 5 for classes 2 and 3. 

```python
#Re-sample the training set

#Find the indexes where ECGs are labeled with certain targets
index0 = np.where(anotaciones[0:nr_train_data]==0)
index1 = np.where(anotaciones[0:nr_train_data]==1)
index2 = np.where(anotaciones[0:nr_train_data]==2)
index3 = np.where(anotaciones[0:nr_train_data]==3)

#Since not all classes are represented an equal amount, I up-sample under-represented classes 2 and 3
newnr_0 = len(index0[0])
newnr_1 = len(index1[0])
mult2 = 20
mult3 = 5
newnr_2 = len(index2[0]) * mult2
newnr_3 = len(index3[0]) * mult3

#Find the new indexes for classes 2 and 3, call them _eq, meaning equalized.
index2_eq = np.repeat(index2[0],mult2)
index3_eq = np.repeat(index3[0],mult3)

print('The amount of signals in the training set that are used per class are now: ', newnr_0,', ', newnr_1,', ', newnr_2,', ', newnr_3)

#Compute the new number of samples in the training set
nr_samples_new = newnr_0 + newnr_1 + newnr_2 + newnr_3

#Define at which index the new label starts in the array
startL1 = newnr_0
startL2 = newnr_0 + newnr_1
startL3 = newnr_0 + newnr_1 + newnr_2

#Fill the array with the corresponding data and labels
data_eq = np.zeros((nr_samples_new, 1206))
data_eq[0:startL1,0:1202] = train_data[index0,:]
data_eq[0:startL1,1202:1206] = train_target[index0,:]

data_eq[startL1:startL2,0:1202] = train_data[index1,:]
data_eq[startL1:startL2,1202:1206] = train_target[index1,:]

data_eq[startL2:startL3,0:1202] = train_data[index2_eq,:]
data_eq[startL2:startL3,1202:1206] = train_target[index2_eq,:]

data_eq[startL3:nr_samples_new,0:1202] = train_data[index3_eq,:]
data_eq[startL3:nr_samples_new,1202:1206] = train_target[index3_eq,:]

#Shuffle training set to prevent that one batch only contains signals of the same class
np.random.shuffle(data_eq)

train_data = data_eq[:,0:1202]
train_target = data_eq[:,1202:1206]
```

I use Keras to define my Neural Net. I started with a network that only contained a dense layer, followed by a non-linear layer, a dense layer and ending with a softmax.
However, this network did not seem to train a lot. Accuracy and loss as a function of epochs did not really update.

A common problem in training neural networks, is the fact that the input of every layer changes
per iteration, since the parameters of the previous layer also changed due to training. Due to this
phenomenon, training can be slow since learning rates have to be set low. Also, parameters need
to be initialized carefully and networks that make use of saturating non-linearities can be hard to
train. 
To solve this problem, I started with changing my non-linear layers to Relu layers.
Compared to a Sigmoid, a Relu has a lower chance of haing gradients of zero during backpropagation, since the positive part of the Relu function is not saturing.

Also Ioffe and Christian Szegedy (2015) proposed a solution to the problem of changing input to every layer per iteration.
The called it batch normalization. A batch normalization layer is added in front of every non-linear layer, which normalizes each layer’s input to unit Gaussian distributions over all dimensions.
After adding these layers to my network, my network started to train a lot better, as I could see in the training graphs that show accuracy and loss over epochs.

After two structures of "dense layer - Batch normalization - Relu", I added a dropout layer to prevent over-fitting. 

Initially I started training the network 10 times, just to see how it trained. Looking at my training graphs (see below), I saw that already after very few epochs overfitting happened. Therefore, I chose to set the number of epochs at 3.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, BatchNormalization
from keras.callbacks import EarlyStopping
import os
import math
import simplejson

batch_size = 32
num_classes = 4
epochs = 3
save_dir = os.path.join(os.getcwd(), 'saved_models')
model_name = 'Finalv1'
saveAccPlots = save_dir + '/AccAndLoss'

input_len = data.shape[1]

# Create model
model = Sequential()
model.add(Dense(256, input_shape=(input_len,)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dense(128))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes))
model.add(Activation('softmax'))


# initiate RMSprop optimizer
opt = keras.optimizers.rmsprop(lr=0.01, decay=1e-6)

# Let's train the model using RMSprop
model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])


# Fit the model on the training data and use validation data
history = model.fit(train_data, train_target,
                        epochs=epochs, verbose = 2,
                        validation_data=(val_data, val_target))



#-------------------------------After training plot the results----------------------------------
# plot history for accuracy
fig1, ax1 = plt.subplots()
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Accuracy per epoch')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.grid()

nrticks = min(epochs, 20)
xvals = [n for n in range(0, epochs, math.ceil(epochs/nrticks))]
plt.xticks( xvals, [n+1 for n in xvals] )

fig1.savefig(saveAccPlots+'/Acc'+model_name + '.png')
plt.close(fig1)

# plot history for loss
fig2 , ax2 = plt.subplots()
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss per epoch')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.grid()

nrticks = min(epochs, 20)
xvals = [n for n in range(0, epochs, math.ceil(epochs/nrticks))]
plt.xticks( xvals, [n+1 for n in xvals] )

fig2.savefig(saveAccPlots+'/Loss'+model_name + '.png')
plt.close(fig2)


#Save all data in a txt file
writeFile = open('AccLoss_'+model_name+'.txt', 'w')
simplejson.dump(history.history, writeFile)
writeFile.close()

#-----------------------------Save model and weights------------------------------------
if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
model_path = os.path.join(save_dir, model_name)
model.save(model_path)
print('Saved trained model at %s ' % model_path)

```

```python
"""
This function prints and plots the confusion matrix.
Normalization can be applied by setting `normalize=True`.
"""
def plot_confusion_matrix(cm,classes=['0','1','2','3'],
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

The following codes plots the non-normalized and normalized confusion matrices for the test-set. 

```python
from sklearn.metrics import confusion_matrix
import itertools

#Predict the scores with the test class
prediction = model.predict(test_data, batch_size=32, verbose=0)

# Set the predictions in the right format to compute the confusion matrix
y_pred = np.argmax(prediction,1)
y_test = anotaciones[(nr_train_data + nr_val_data):nr_samples]

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, normalize=True,
                      title='Normalized confusion matrix')

plt.show()

#Compute final score for class
score = 0.9*(cnf_matrix[1,0] + cnf_matrix[2,0] + cnf_matrix[3,0]) + 0.1*(cnf_matrix[0,1]+cnf_matrix[0,2]+cnf_matrix[0,3])+ 0.2*(cnf_matrix[1,2]+cnf_matrix[1,3]+cnf_matrix[2,1]+cnf_matrix[2,3] + cnf_matrix[3,1]+cnf_matrix[3,2])
print('Final score of model: ', score)
```

## Score y ganador

La cátedra se reservó parte del dataset para evaluar los modelos y el puntaje de cada modelo será calculado de la siguiente forma:

score = 0.9 * Cantidad de latidos con arritmias tomados como normales + 0.1 * Cantidad de latidos normales tomados como latidos con arritmias + 0.2 * Cantidad de latidos con arritmias de cierto tipo tomados como latidos con una arritmia diferente

El alumno que obtenga el menor score será el ganador del challenge.
En el extramadamente improbable caso de haber un empate no habrá ganador.
